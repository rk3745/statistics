```python
question-- one of the dataset incorporating missing values is given as[5,8,None,7,10,None,6].
write a python program to find out the mean and median by handling the miss values. do not use function concept or inbuilt function.
requirement-- your program should print answers in format of
1. original data
2. handled data
3. mean
4. median
```

```python
data=[5,8,None,7,10,None,6]
s= 0
count=0
for x in data:
  if x is not None:
    s+=x
  count+=1
mean_val=s/count
inputed = []
for x in data:
  if x is not None:
    inputed.append(mean_val)
  else:
      inputed.append(x)
print("original data:",data)
print("handled data:",inputed)
print("mean:",mean_val)
median = (data[3]+data[4])/2
print("median:",median)

```
