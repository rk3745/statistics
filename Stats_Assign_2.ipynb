***Ques_01 = 2 fair and unbiased dice are rolled simultaneously to get some outcomes during the
experiment. Write a python program to compute number of outcomes and probability of getting both
faces same.***

```python
# Total possible outcomes for two dice
total_outcomes = 6 * 6

# Favorable outcomes: both dice show same face (1-1, 2-2, ..., 6-6)
favorable_outcomes = 6

# Probability calculation
probability = favorable_outcomes / total_outcomes

# Display results
print(f"Number of total outcomes: {total_outcomes}")
print(f"Number of favorable outcomes: {favorable_outcomes}")
print(f"Probability of both faces same: {probability:.4f} or 1/6")
```
