***Ques_01 = Write a python program to compute accuracy, precision, recall, F1-score for the
confusion matrix given below.***

```python
# Confusion matrix values
tp = 199  # True Positives (Pos predicted as Pos)
fn = 1    # False Negatives (Pos predicted as Neg)
fp = 3    # False Positives (Neg predicted as Pos)
tn = 247  # True Negatives (Neg predicted as Neg)

# Compute metrics
total = tp + tn + fp + fn
accuracy = (tp + tn) / total
precision = tp / (tp + fp)
recall = tp / (tp + fn)
f1_score = 2 * (precision * recall) / (precision + recall)

# Display results
print(f"Accuracy:  {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1-Score:  {f1_score:.4f}")
```
