```python
# Grouped data
classes = [(15,25), (25,35), (35,45), (45,55), (55,65)]
frequencies = [350, 375, 120, 80, 90]

# Step 1: Calculate midpoints
midpoints = [(low + high) / 2 for low, high in classes]

# Step 2: Calculate Mean
N = sum(frequencies)
mean = sum(f * x for f, x in zip(frequencies, midpoints)) / N

# Step 3: Calculate Cumulative Frequencies
cumulative_frequencies = []
cumulative = 0
for f in frequencies:
    cumulative += f
    cumulative_frequencies.append(cumulative)

# Step 4: Find Median
median_position = N / 2

for i, cf in enumerate(cumulative_frequencies):
    if cf >= median_position:
        median_class_index = i
        break

L = classes[median_class_index][0]  # Lower boundary of median class
F = cumulative_frequencies[median_class_index - 1] if median_class_index != 0 else 0
f = frequencies[median_class_index]
h = classes[0][1] - classes[0][0]  # Class width

median = L + ((median_position - F) / f) * h

# Step 5: Find Mode
modal_class_index = frequencies.index(max(frequencies))
L_mode = classes[modal_class_index][0]
f1 = frequencies[modal_class_index]
f0 = frequencies[modal_class_index - 1] if modal_class_index != 0 else 0
f2 = frequencies[modal_class_index + 1] if modal_class_index != len(frequencies) - 1 else 0

mode = L_mode + ((f1 - f0) / ((2 * f1) - f0 - f2)) * h

# Step 6: Print Results
print("Mean  ≈", round(mean, 2))
print("Median ≈", round(median, 2))
print("Mode  ≈", round(mode, 2))
```
